```python
yes = data.loc[data.dec==1]
yes
```

```python
no = data.loc[data.dec==0]
no
```

```python
yes_dec= yes.income
yes_dec
```

```python
no_dec= no.income
no_dec
```

```python
yes.income.mean()
```

```python
no.income.mean()
```

```python
t_value, p_value = stats.ttest_ind(yes_dec, no_dec)
print("t_value : ",t_value,"\np_value : ", p_value)
alpha = 0.05
```

```python
if p_value<alpha:
    print("Reject the null hypothesis . the dating decision is based on high income")
else:
        print("fail to reject the null hypothesis. the dating decision is not based on high income")
```

```python
fun_less=no.fun
fun_less
```

```python
fun_more=yes.fun
fun_more
```

```python
t_value, p_value = stats.ttest_ind(fun_more, fun_less)
print("t_value : ",t_value,"\np_value : ", p_value)
alpha = 0.05
```

```python
if p_value<alpha:
    print("Reject the null hypothesis . the dating decision is based on funny character")
else:
        print("fail to reject the null hypothesis. the dating decision is not based on funny")
```
